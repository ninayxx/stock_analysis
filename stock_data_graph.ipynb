```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
sns.set(style="darkgrid")

df_nvda = pd.read_csv("NVDIA_Stock_Data.csv", parse_dates=["Date"], index_col="Date")
df_amd = pd.read_csv("AMD_Stock_Data.csv", parse_dates=["Date"], index_col="Date")

df_nvda = df_nvda.sort_index()
df_amd = df_amd.sort_index()

combined = pd.DataFrame({
    "NVIDIA": df_nvda["Close"],
    "AMD": df_amd["Close"]
}).dropna()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(combined.index, combined["NVIDIA"], label="NVIDIA Close Price")
plt.plot(combined.index, combined["AMD"], label="AMD Close Price")
plt.xlabel("Time")
plt.ylabel("Price (USD)")
plt.title("NVIDIA vs AMD Stock Prices Over Time")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(combined["NVIDIA"], combined["AMD"], alpha=0.6)
corr = combined.corr().iloc[0,1] #Reference pearson_correlation.py for the extended calculation
plt.xlabel("NVIDIA Close Price (USD)")
plt.ylabel("AMD Close Price (USD)")
plt.title(f"Scatter Plot of NVIDIA vs AMD Stock Prices\nPearson Correlation: {corr:.3f}")
plt.grid(True)
plt.show()
```

```python
returns = combined.pct_change().dropna()
plt.figure(figsize=(10, 6))
sns.histplot(returns["NVIDIA"], bins=50, kde=True, color="blue", label="NVIDIA Returns")
plt.xlabel("Daily Returns")
plt.ylabel("Frequency")
plt.title("Histogram of Daily Returns for NVIDIA")
plt.tight_layout()
plt.show()
```

```python
returns = combined.pct_change().dropna()
plt.figure(figsize=(10, 6))
sns.histplot(returns["AMD"], bins=50, kde=True, color="orange", label="AMD Returns")
plt.xlabel("Daily Returns")
plt.ylabel("Frequency")
plt.title("Histogram of Daily Returns for AMD")
plt.tight_layout()
plt.show()
```
